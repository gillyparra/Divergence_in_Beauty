```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import time
import csv
import json
from itertools import chain



# U20 wc called the youth wc for these years 
years = [1977, 1979, 1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999, 2001, 2003, 2005]

# Initialize an empty list to store DataFrames
dfs = []

for year in years:
    # Construct the URL
    url = f'https://en.wikipedia.org/wiki/{year}_FIFA_World_Youth_Championship'
    
    # Fetch the page
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the last "wikitable" on the page
    table = soup.find_all('table', {'class': "wikitable"})[-1]
    
    # Read the table into a DataFrame
    df_year = pd.read_html(str(table))[0]
    
    # Add a column for the year
    df_year['Year'] = year
    
    # Append the DataFrame to the list
    dfs.append(df_year)

# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(dfs, ignore_index=True)

# Drop rows where the 'Team' column is NaN
final_df.dropna(subset=['Team'], inplace=True)

# Drop the 'Final result' column
final_df.drop(columns=['Final result'], inplace=True)

# Drop the 'GD' column
final_df.drop(columns=['GD'], inplace=True)

# Display the first few rows of the final DataFrame
print(final_df.head())


```

```python
# You can then save this DataFrame to a CSV file if needed
final_df.to_csv('C:/Users/Guill/Downloads/U20 77-05.csv', index=False)
```

```python
import requests
import json
import pandas as pd
from itertools import chain

# Modify the headers to include a User-Agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

# Initialize an empty list to store formatted data from all rankings
all_rankings_formatted_data = []

# Combine the two ranges into a single iterable
id_ranges = chain(range(1, 154), range(8000, 14501))

for rank_id in id_ranges:
    url = f'https://www.fifa.com/api/ranking-overview?locale=en&dateId=id{rank_id}'
    
    response = requests.get(url, headers=headers)
    
    # Print progress
    print(f"Fetching rank id: {rank_id}")
    
    if response.status_code == 200:
        data = json.loads(response.text)
        
        # Extract ranking data
        rankings_list = data.get('rankings', [])
        
        # Check if rankings_list is not empty
        if rankings_list:
            # Process each ranking
            for ranking_data in rankings_list:
                all_rankings_formatted_data.append({
                    'dateId': f'id{rank_id}',  # Dynamic date ID
                    'country': ranking_data['rankingItem']['name'],
                    'rank': ranking_data['rankingItem']['rank'],
                    'previousRank': ranking_data['rankingItem']['previousRank'],
                    'totalPoints': ranking_data['rankingItem']['totalPoints'],
                    'previousPoints': ranking_data['previousPoints'],
                    'flagUrl': ranking_data['rankingItem']['flag']['src'],
                    'countryUrl': 'https://www.fifa.com' + ranking_data['rankingItem']['countryURL'],
                    'confederation': ranking_data['tag']['text'],
                })
        else:
            print("No rankings data found for this rank id.")
    else:
        print(f"Failed to fetch data for rank id {rank_id}. Status code:", response.status_code)
    
    
# Convert the collected data into a DataFrame
rankings_df = pd.DataFrame(all_rankings_formatted_data)

```
